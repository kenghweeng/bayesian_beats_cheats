```python
import pandas as pd
from pyvis.network import Network
```

```python
max_edge_weights = pd.read_csv("../data/max_edge_weights.csv")
max_edge_weights.head()
```

```python
unified_node_data = pd.read_csv("../data/unified_node_data.csv")
unified_node_data['label'] = unified_node_data.apply(lambda row: 0 if pd.isna(row["confessed_assignments"]) else 1, axis=1)
unified_node_data.head()
```

```python
net = Network(height="700px", width="100%")
```

```python
def get_hover(df):
    return df.apply(lambda row: '<strong>Name</strong>: ' + str(row['name']) + '<strong>Faculty</strong>: ' + str(row['faculty']) + '<br><strong>Admit Year</strong>: ' + str(row['admit_year']) + '<br><strong>Programme</strong>: ' + str(row['programme']), axis = 1)
```

```python
def get_color(df):
    return df.apply(lambda row: 'red' if row['label'] == 1 else '#d9f7f7', axis = 1)
```

```python
net.add_nodes(unified_node_data['name'], label=unified_node_data['name'], title=get_hover(unified_node_data), color=get_color(unified_node_data))
```

```python
for index, row in max_edge_weights.iterrows():
    try:
        net.add_edge(row['NodeID1'], row['NodeID2'], value=row['edge_weights'], title=row['edge_weights'], physics=False)
    except AssertionError as err:
        print(err)
        pass
```

```python
net.show("bbc.html")
```
